In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm

avg = df.mean(axis=1)
se = df.std(axis=1) / (len(df.columns))**0.5               # Standard Error
conf_int = 1.96 * se                                       # Confidence interval

hlines = []                                                # To store the 2 horizontal lines positions

#plt.figure()
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
bars = plt.bar(range(len(avg)), avg, yerr=conf_int, color='#4d94ff')
plt.xticks(range(len(avg)), df.index)
plt.title('Votes per year', fontsize=15)
colors = plt.get_cmap('RdBu')
normalize = mpl.colors.Normalize(vmin=0, vmax=2)

def onclick(event):
    global hlines
    global bars
    
    # If first click, add a horizontal line where clicked
    if len(hlines) == 0:
        hlines.append(event.ydata)
        plt.axhline(y=event.ydata, color='r', linestyle='-')
    
    # If second click, add the second horizontal line and color the area between the two lines
    elif len(hlines) == 1:
        hlines.append(event.ydata)
        hlines.sort()
        range_up = hlines[1]
        range_down = hlines[0]
        plt.axhline(y=event.ydata, color='r', linestyle='-')
        plt.axhspan(range_down, range_up, facecolor='y', alpha=0.5)

        # Then color each bar in the plot according to its overlap with the selected range
        for i, bar in enumerate(bars):
            lim_up = bar.get_height() + conf_int.iloc[i]             # Higher edge of the bar
            lim_down = bar.get_height() - conf_int.iloc[i]           # Lower edge of the bar

            # If selected range is too high, make the bar red (0% probability to happen)
            if range_down > lim_up:
                bar.set_color(colors(normalize(0)))
                
            # If selected range is too low, make the bar blue (0% probability to happen)
            elif range_up < lim_down:
                bar.set_color(colors(normalize(2)))
            
            # If selected range include the whole bar range, make the bar white (100% probability to happen)
            elif range_up > lim_up and range_down < lim_down:
                bar.set_color(colors(normalize(1)))
            
            # If selected range is completely included inside the bar range,
            # calculate the overlap area and color the bar accordingly
            elif range_up < lim_up and range_down > lim_down:
                overlap = (range_up - range_down) / (lim_up - lim_down)
                bar.set_color(colors(normalize(overlap)))
                
            # If selected range is partially included inside the bar range
            # but slightly higher, calculate the overlap area and color the
            # bar accordingly
            elif range_down < lim_up < range_up:
                overlap = (lim_up - range_down) / (lim_up - lim_down)
                bar.set_color(colors(normalize(overlap)))
            
            # If selected range is partially included inside the bar range
            # but slightly lower, calculate the overlap area and color the
            # bar accordingly
            elif range_up > lim_down > range_down:
                overlap = (range_up - lim_down) / (lim_up - lim_down)
                bar.set_color(colors(normalize(overlap)))

    # If third click, reset figure and plot again orignal data
    else:
        hlines = []
        plt.cla()
        bars = plt.bar(range(len(avg)), avg, yerr=conf_int, color='#4d94ff')
        plt.xticks(range(len(avg)), df.index)
        plt.title('Votes per year', fontsize=15)

scalarmappaple = cm.ScalarMappable(norm=normalize, cmap=colors)
scalarmappaple.set_array(np.arange(-1, 1, 0.1))
cbar = plt.colorbar(scalarmappaple, ticks=[0, 1, 2])
cbar.ax.set_yticklabels(['0', '1', '0'])
plt.show()

_ = plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```
